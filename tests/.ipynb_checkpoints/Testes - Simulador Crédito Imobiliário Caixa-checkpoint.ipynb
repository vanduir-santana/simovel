```python
import urllib.request
import urllib.parse
import unidecode
import ngram
```

```python
def remover_acentos(s, maiusc=True):
    palavra_acentuada = s.upper() if maiusc else s
    return unidecode.unidecode(palavra_acentuada)

palavra_acentuada = 'Oração'
print('acentuada: {}'.format(palavra_acentuada))
print('sem acento: {}'.format(remover_acentos(palavra_acentuada)))
```

```python
url = 'http://www8.caixa.gov.br/siopiinternet-web/dwr/call/plaincall/SIOPIAjaxFrontController.callActionForwardMethodLista.dwr'
#url2 = 'http://www8.caixa.gov.br/siopiinternet-web/simulaOperacaoInternet.do?method=validarMunicipioFeirao&cidade=2236&requisicaoModal=true'
url2 = 'http://www8.caixa.gov.br/siopiinternet-web/simulaOperacaoInternet.do?method=validarMunicipioFeirao&cidade={cod_cidade}&requisicaoModal=true'
```

```python
UFS = ('AC', 'AL', 'AP', 'AM', 'BA', 'CE', 'ES', 'GO', 'MA', 'MT', 
       'MS', 'MG', 'PA', 'PB', 'PR', 'PE', 'PI', 'RJ', 'RN', 'RS', 
       'RO', 'RR', 'SC', 'SP', 'SE', 'TO', 'DF')
```

```python
def ler_uf():
    uf = ''
    while True:
        uf = input("Entre com a UF: ")
        uf = uf.upper()
        if uf in UFS:
            #break
            return uf
        else:
            print('Favor digitar uma UF válida!')
```

```python
def obter_cidades(uf):
    params =  {
        'callCount': 1,
        'page': '/siopiinternet-web/simulaOperacaoInternet.do?method: inicializarCasoUso',
        'httpSessionId': '8xcvLr-SmEs-yq6RtTF5wXY6.habitacao_bbrnpapllx013:siopi-internet-prd-node01_lx013',
        'scriptSessionId': '99E63DC129315328D07CCF8C813117E6517',
        'c0-scriptName': 'SIOPIAjaxFrontController',
        'c0-methodName': 'callActionForwardMethodLista',
        'c0-id': 0,
        'c0-param0': 'string:%2FsimulaOperacaoInternet',
        'c0-param1': 'string:listarCidades',
        'c0-param2': 'string:uf%3D{}'.format(uf),
        'batchId': 2
    }

    campos = urllib.parse.urlencode(params)
    dados = campos.encode('utf-8')
    with urllib.request.urlopen(url, dados) as response:
        response_text = response.read()
        r = response_text.decode('utf-8')
        return r
```

### Extrair dicionário da string de retorno de cidades da caixa

```python
#class Cidade(object):
#    codigo = 0
#    nome = ''
#    codigo = 2090
#    ehFeirao = False
#    nome = "ABADIA DE GOIAS"
#    nomeSemAspa = "ABADIA DE GOIAS"
#    nomeUF = "ABADIA DE GOIAS - GO"
#    siglaUf = "GO"
#    valorBasePaginacao = 2090
    
ACHAR1='s0.codigo='
ACHAR2='dwr.engine.'
CODIGO=0
NOME=13
NOME_SEM_ASPA=15

def extrair_cidades(s):
    '''
    Extrai cidades e seus codigos da chamada a url retornada em variáveis javascript.
    '''
    pos = s.find(ACHAR1)
    print('Pos: {}'.format(pos))
    if pos < 50:
        print('Não é possível continuar pois houve algum erro ao retornar as cidades!')
        return False

    # pega a partir do começo da declaração das variáveis
    s = s[pos:]

    d = {}
    # varrer linhas
    for linha in s.split('\n'):
        # verificar se chegou ao fim
        if linha.startswith(ACHAR2):
            break

        # retirar primeira parte, o objeto (antes do primeiro ponto)
        pos = linha.find('.')
        texto = linha[:pos + 1]
        linha = linha.replace(texto, '')

       # separar campos
        campos = linha.split(';')
        nome = campos[NOME].split('=')[1].replace('"', '').replace('\\', '')
        nome_sem_aspa = campos[NOME_SEM_ASPA].split('=')[1].replace('"', '')
        codigo = campos[CODIGO].split('=')[1]

        #d[nome_sem_aspa] = int(codigo)
        d[nome_sem_aspa] = (codigo, nome)
    
    return d
```

### Main1

```python
print('Lendo UF...')
uf = ler_uf()
print('Obtendo cidade...')
r = obter_cidades(uf)
print('Extraindo cidades do modo raw')
d = extrair_cidades(r)
```

```python
def procurar(t, l, max_res=100):
    '''
    Procurar texto (t) por similaridade em uma lista (l).
    '''
    pesq = ngram.NGram(l)
    print('Texto, Similaridade')
    cont = 0
    for j in pesq.search(t, threshold=0.1):
        print('{}, {}'.format(j[0], j[1]))
        # texto eh igual
        if j[1] == 1.0:
            print('Similaridade Igual')
            break
        
        if cont == max_res:
            break
            
        cont += 1
        

l2 = [i[1] for i in d.values()]
while True:
    print('*' * 100)
    t = input("Entre com o nome da cidade ou 0 pra sair: ")
    if t == '0':
        break
    print('*' * 100)
    t = remover_acentos(t)
    procurar(t.upper(), l2, max_res=10)
```
